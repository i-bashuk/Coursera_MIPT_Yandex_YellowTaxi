```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
```

### 1. Загрузите данные за май 2016 года. 

```python
# Константы
long = [-74.25559, -73.70001]
lat = [40.49612, 40.91553]
ESB = [-73.985428, 40.748817]
ESB_zone = int((ESB[0] - long[0]) // ((long[1] - long[0]) / 50) * 50 + (ESB[1] - lat[0]) // ((lat[1]-lat[0]) / 50))
#region = pd.read_csv('regions.csv', ';')
```

```python
# Загрузим данные
df = pd.read_csv('/mnt/e/python/taxi/yellow_tripdata_2016-05.csv', 
                 parse_dates=['tpep_pickup_datetime', 'tpep_dropoff_datetime'],
                 usecols=range(7))
df[:3]
```

### 2. Проведите фильтрацию.

```python
# Удаляем поездки с нулевой длительностью, нулевым количеством пассажиров и нулевым расстоянием поездки по счётчику 
df = df[(df.tpep_pickup_datetime != df.tpep_dropoff_datetime) & (df.trip_distance > 0) & (df.passenger_count > 0)]
# Удаляем поездки скоординатами начала за пределами Нью-Йорка
df = df[(long[0]<=df.pickup_longitude) & (df.pickup_longitude<=long[1])&
        (lat[0]<=df.pickup_latitude) & (df.pickup_latitude<=lat[1])]
# Округляем время начала поездки до целого часа
df['tpep_pickup_datetime'] = df.tpep_pickup_datetime.dt.floor(freq='H')
```

### 3. Агрегируйте данные за май 2016 года по часам и областям.

```python
def foo(df):
    x = df.pickup_longitude.values
    y = df.pickup_latitude.values
    ret = stats.binned_statistic_2d(x, y, None, 'count', bins=50, range=[long, lat])
    bdf = pd.DataFrame(ret.statistic.ravel())
    bdf.index = bdf.index + 1
    return bdf.transpose()
```

```python
%%time
dfg = df.groupby('tpep_pickup_datetime').apply(foo)
dfg.index = dfg.index.droplevel(1)
```

### 4. Выведите первые строки полученной таблицы с указанием времени, идентификатора зоны и количества поездок.

### 5. Постройте график количества поездок жёлтого такси из ячейки, содержащей Empire State Building.  

```python
#  График количества поездок жёлтого такси из ячейки, содержащей Empire State Building.
plt.figure(figsize=(20, 7))
ESBdf = dfg.iloc[:, ESB_zone]
plt.plot(ESBdf.index, ESBdf.values)
plt.title('Zone with Empire State Building', fontsize=24)
plt.xlabel('Days/Hours', fontsize=16)
plt.ylabel('Trips', fontsize=16);
```
