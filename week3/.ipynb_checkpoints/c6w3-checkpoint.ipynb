```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
```

```python
for d in pd.date_range('2015-03', '2016-03', freq='M'):
    print(f'aaa{d.year}_{d.month:02}')
```

```python
def load_pickles(start, end, zone):
    """
    format YYYY-MM, start included, end - not
    """
    res = pd.Series(dtype=np.float64)
    for d in pd.date_range(start, end, freq='M'):
        with open(os.path.join(PATH_TO_DATA, f'pickles/yellow_tripdata_{d.year}-{d.month:02}.pkl'), 'rb') as f:
            df = pickle.load(f)
            #print(type(df[zone]))
        res = res.append(df[zone])
    return res
            
```

```python
def add_feature(T, K, season = 168):
    res=np.empty((T, 2*K))
    arr = 2 * np.pi * np.outer(np.arange(1, T+1), np.arange(1, K+1)) / season
    res[:, :K] = np.sin(arr)
    res[:, K:] = np.cos(arr)
    return res
```

```python
y = load_pickles('2014-05', '2016-06', 1282)
y_day = y.resample('D').sum()
```

```python
plt.figure(figsize=(15,7))
plt.plot(y.resample('D').sum())
```

```python
plt.figure(figsize=(15,10))
sm.tsa.seasonal_decompose(y_day).plot()
```

```python
k_ = [2, 5, 10, 20, 50, 100]
fig, axes = plt.subplots(3, 2, figsize=(15,7))
for i, k in enumerate(tqdm(k_)):
    X = np.hstack((add_feature(y.shape[0], k), add_feature(y.shape[0], k, season=8766)))
    lr = LinearRegression()
    lr.fit(X, y)
    res = pd.Series(lr.predict(X), index=y.index).resample('D').sum()
    axes[i//2, i%2].plot(y_day.index, y_day, 'r')
    axes[i//2, i%2].plot(res, 'b--')
    print(mean_squared_error(y_day, res))
```

```python
k=5
trend = (np.arange(1, y.shape[0]+1)[:, np.newaxis])
X = np.hstack((add_feature(y.shape[0], k), add_feature(y.shape[0], k, season=8766), trend))
lr = LinearRegression()
lr.fit(X, y)
res = pd.Series(lr.predict(X), index=y.index).resample('D').sum()
plt.figure(figsize=(15,7))
plt.plot(y_day.index, y_day, 'r')
plt.plot(y_day.index, res, 'b--')
print(mean_squared_error(y_day, res))
```
