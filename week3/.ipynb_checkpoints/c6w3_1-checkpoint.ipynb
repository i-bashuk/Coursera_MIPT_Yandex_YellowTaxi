```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
from itertools import product
import warnings
```

```python
def load_pickles(start, end, zone):
    """
    format YYYY-MM, start included, end - not
    """
    res = pd.Series(dtype=np.float64)
    for d in pd.date_range(start, end, freq='M'):
        with open(os.path.join(PATH_TO_DATA, f'pickles/yellow_tripdata_{d.year}-{d.month:02}.pkl'), 'rb') as f:
            df = pickle.load(f)
            #print(type(df[zone]))
        res = res.append(df[zone])
    return res
            
```

```python
def add_feature(T, K, season = 168):
    res=np.empty((T, 2*K))
    arr = 2 * np.pi * np.outer(np.arange(1, T+1), np.arange(1, K+1)) / season
    res[:, :K] = np.sin(arr)
    res[:, K:] = np.cos(arr)
    return res
```

## Подбор модели

Посмотрим на ACF и PACF полученного ряда:

```python
resid_diff = resid.diff(24).diff()[25:]
plt.figure(figsize=(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(resid_diff.values.squeeze(), lags=168, ax=ax)
plt.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(resid_diff.values.squeeze(), lags=168, ax=ax)
plt.show()
```

Начальные приближения: Q=0, q=5, P=1, p=5

```python
ps = range(1, 2)
d=1
qs = range(1, 2)
Ps = range(1, 2)
D=1
Qs = range(1, 2)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(y, X, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 24),
                                        enforce_invertibility=False, enforce_stationarity=False).fit(disp=False)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Лучшая модель:

```python
print(best_model.summary())
```

Её остатки:

```python
from scipy import stats
plt.figure(figsize=(15,8))
plt.subplot(211)
best_model.resid.plot(title='Остатки модели')
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid.values.squeeze(), lags=48, ax=ax, title='Autocorrelation')

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid, 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid)[1])
```

Остатки несмещены (подтверждается критерием Стьюдента) стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается критерием Льюнга-Бокса (Prob(Q) = 0.95) и коррелограммой).
